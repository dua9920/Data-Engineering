```python
# 각각의 데이터를 가져온다.
# csv, txt 등의 파일은 read_csv()로 가져올 수 있다.
flights_df_raw = pd.read_csv('data/flights.csv')
airports_df = pd.read_csv('data/airports.csv')
airlines_df = pd.read_csv('data/airlines.csv')
```

```python
# flight 데이터의 정보를 가져온다.
# 열의 이름과 타입을 알 수 있다.
# 총 5819079 개의 데이터 : 0 ~ 5819078 INDEX
# 총 31개의 column
flights_df_raw.info()
```

```python
# 해당 전처리를 하기 위하여 필요한 데이터만 남겨둔다.
flights_df_raw = flights_df_raw.iloc[:, 0:12]
# flights_df_raw = flights_df_raw.drop(['DAY_OF_WEEK'], axis=1)
flights_df_raw = flights_df_raw.drop(columns=['DAY_OF_WEEK', 'TAIL_NUMBER', 'DEPARTURE_TIME'])
flights_df_raw.info()
```

```python
# SCHEDULED_DEPARTURE 열의 시간을 timestamp 형식으로 표현
# ex) 2015-01-01 00:05:00
# ex) 만약 103이라면 2015-01-01 01:03:00
def convert_to_timedate(x, col='SCHEDULED_DEPARTURE'):
    return (
        pd.to_datetime(
            x['YEAR'].astype(str) + '-' + x['MONTH'].astype(str) + '-' + x['DAY'].astype(str) +
            ' ' + x.assign(zero='0')['zero'] * (4 - x[col].astype(str).str.len()) +
            x[col].astype(str), format='%Y-%m-%d %H%M', errors='coerce'
        )
    )
```

```python
# flights_df_raw 데이터 정제화
flights_df = (
    flights_df_raw
     # 행을 기준으로 NaN 값이 있는 데이터를 삭제한다.
    .dropna()
    # ORIGIN_AIRPORT 값이 BOS, SFO, LAX, JFX 이고
    # DEPARTURE_DELAY 가 24*60 보다 작다면 남긴다.
    [lambda x:
         (x['ORIGIN_AIRPORT'].isin(['BOS', 'SFO', 'LAX', 'JFK'])) &
         (x['DEPARTURE_DELAY'] <= 24*60)
    ]
    # 열을 할당한다.
    # FLIGHT_NUMBER : str 값으로 바꾼다.
    # SCHEDULED_DEPARTURE : TimeStamp 형식으로 변환.
    # IS_DELAYED :  15분 이상 지연이 있었던 값을 선택한다.
    .assign(
        FLIGHT_NUMBER=lambda x: x['FLIGHT_NUMBER'].astype(str),
        SCHEDULED_DEPARTURE=lambda x: convert_to_timedate(x),
        IS_DELAYED=lambda x: np.select(
            [x['DEPARTURE_DELAY'] >= 15], [1],
            default=0
        )
    )
    # YEAR, MONTH, DAY 열은 삭제한다. (TimeStamp로 이미 변환하였으므로)
    .drop(['YEAR', 'MONTH', 'DAY'], axis=1)
)
```

### 2. 조인, 집계 (Joins, aggregations)
----

```python
# 공항 이름 가져오기
# flights_df 와 airports_df 를 ORIGIN_AIRPORT 기준으로 병합한다.
result = (
  flights_df
  .merge(airports_df, left_on='ORIGIN_AIRPORT', right_on='IATA_CODE')
)
result.head()
```

```python
# 공항 당 항공편 수 확인하기
result = (
  flights_df
  # ORIGIN_AIRPORT 열을 기준으로 데이터를 그룹화한다.
  # 즉, BOS, SFO, LAX, JFK 의 데이터를 모으는 것이다.
  .groupby(['ORIGIN_AIRPORT'])
  # 그룹화 한 것들을 집계한다. (행의 갯수를 계산)
  .apply(lambda x: pd.Series({
      'nb_flights': len(x['ORIGIN_AIRPORT'])
  }))
  .reset_index()
)
result.head()
```

```python
# 대형 항공사의 지연된 항공편 확인하기
result = (
  flights_df
  # AIRLINE 열을 기준으로 그룹화한다.
  .groupby(['AIRLINE'])
  # nb_flights : FLIGHT_NUMBER 의 집계값을 저장.
  # perc_delayed : IS_DELAYED 의 평균값을 저장.
  .apply(lambda x: pd.Series({
      'nb_flights': len(x['FLIGHT_NUMBER']),
      'perc_delayed': round(np.average(x['IS_DELAYED']), 4)
  }))
  .reset_index()
  # 내림차순으로 
  .assign(rank_size=lambda x: x['nb_flights'].rank(method='first', ascending=False))
  [lambda x: x['rank_size'] <= 10]
  .drop('rank_size', axis=1)
  # Display airline name
  .merge(airlines_df, left_on='AIRLINE', right_on='IATA_CODE')
  .reset_index(drop=True)
)
result.head()
```

```python
# 대형 항공사의 지연된 항공편 확인하기
result = (
  flights_df
  # AIRLINE 열을 기준으로 그룹화한다.
  .groupby(['AIRLINE'])
  # nb_flights : FLIGHT_NUMBER 의 집계값을 저장.
  # perc_delayed : IS_DELAYED 의 평균값을 저장.
  .apply(lambda x: pd.Series({
      'nb_flights': len(x['FLIGHT_NUMBER']),
      'perc_delayed': round(np.average(x['IS_DELAYED']), 4)
  }))
  .reset_index()
  # nb_flights 집계값을 기준으로 순위를 매긴다.
  .assign(rank_size=lambda x: x['nb_flights'].rank(method='first', ascending=False))
  # rank_size가 10 이하인 값들만 남겨둔다.
  [lambda x: x['rank_size'] <= 10]
  # rank_size 열을 삭제한다.
  .drop('rank_size', axis=1)
  # airlines_df 데이터와 병합한다.
  # AIRLINE 열이 중복되어 _x, _y 값으로 나뉘어 진다.
  # AIRLINE_y 에 풀 이름이 들어간다.
  .merge(airlines_df, left_on='AIRLINE', right_on='IATA_CODE')
  # 행 인덱스 초기화
  .reset_index(drop=True)
)
result
```

### 3. 데이터 프레임 변환 (Data frame transformations)
----

```python
# 공항 별 지연의 시간적 변화 확인하기
result = (
    flights_df
    # MONTH : SCHEDULED_DEPARTURE 열을 사용하여 월을 나타내는 열을 생성.
    # 즉, March, December 이런 식으로 월을 표현.
    # MOTN_NUM : SCHEDULED_DEPARTURE 열을 사용하여 월을 슷자로 나타내는 열을 생성.
    # 즉, 3, 12 이런 식으로 월을 표현.
    .assign(
        MONTH=lambda x: x['SCHEDULED_DEPARTURE'].dt.strftime('%B'),
        MONTH_NUM=lambda x: x['SCHEDULED_DEPARTURE'].dt.strftime('%m').astype(int)
    )
    # ORIGIN_AIRPORT, MONTH, MONTH_NUM 을 그룹화한다.
    .groupby(['ORIGIN_AIRPORT', 'MONTH', 'MONTH_NUM'])
    # IS_DELAYED 열을 기준으로 평균을 구하여 perc_delayed 을 값을 집계한다. 
    .apply(lambda x: pd.Series({
        'perc_delayed': round(np.average(x['IS_DELAYED']), 4)
    }))
    # 행 인덱스 초기화.
    .reset_index()
    # Data Frame을 모양을 재정비한다.
    # ORIGIN_AIRPORT 을 기준으로 각 월에 따른 perc_delayed 값을 나타낸다. 
    .pivot_table(
        columns='ORIGIN_AIRPORT',
        values='perc_delayed',
        index=['MONTH', 'MONTH_NUM'],
        aggfunc=np.sum
    )
    # 열을 지우고 행 인덱스 초기화.
    .rename_axis(None, axis=1)
    .reset_index()
    # MONTH_NUM을 기준으로 정렬한다.
    .sort_values(['MONTH_NUM'])
    # MONTH_NUM 열을 삭제한다.
    .drop(['MONTH_NUM'], axis=1)
    # 행 인덱스 초기화.
    # 결론적으로, 월을 기준으로 정렬되어 perc_delayed 집계값이 나타내진다.
    # 월별로 지연된 퍼센트를 확인할 수 있다.
    .reset_index(drop=True)
)
result
```

```python
# 매월 항공편 변경률 확인하기
result = (
    flights_df
    # MONTH, MONTH_NUM : SCHEDULED_DEPARTURE 데이터를 사용하여 월을 나타낸다.
    .assign(
        MONTH=lambda x: x['SCHEDULED_DEPARTURE'].dt.strftime('%B'),
        MONTH_NUM=lambda x: x['SCHEDULED_DEPARTURE'].dt.strftime('%m').astype(int)
    )
    # ORIGIN_AIRPORT, MONTH, MONTH_NUM 을 그룹화한다.
    .groupby(['ORIGIN_AIRPORT', 'MONTH', 'MONTH_NUM'])
    # nb_flights : FLIGHT_NUMBER 의 갯수를 세어 집계한다.
    .apply(lambda x: pd.Series({
        'nb_flights': len(x['FLIGHT_NUMBER'])
    }))
    # 행 인덱스를 초기화하고, ORIGIN_AIRPORT, MONTH_NUM 순을 기준으로 정렬한다.
    # 각 항공사마다 월 별로 정렬이된다. 
    .reset_index()
    .sort_values(['ORIGIN_AIRPORT', 'MONTH_NUM'])
    # 어질러진 인덱스를 다시 초기화한다.
    .reset_index(drop=True)
    # nb_flights_before : ORIGIN_AIRPORT 을 기준으로 그룹핑한 값에 이 전 값을 넣는다.
    # 즉, 2월의 nb_flights_before 값은 1월의 nb_flights 값이 된다.
    # perc_chance : 차이의 값을 비교하여 나타낸다.
    .assign(
        nb_flights_before=lambda x:
            x.groupby(['ORIGIN_AIRPORT'])['nb_flights'].shift(1),
        perc_chance=lambda x:
            round((x['nb_flights'] - x['nb_flights_before']) / x['nb_flights_before'], 2)
    )
    # MONTH_NUM 열을 삭제한다.
    # 결론적으로, 매월마다의 항공편 변화률을 확인할 수 있다.
    .drop(['MONTH_NUM'], axis=1)
)
result.head(10)
```

```python
# 출발지 공항 별 가장 인기있는 노선 확인하기.
result = (
    flights_df
    # ORIGIN_AIRPORT, DESTINATION_AIRPORT 열을 기준으로 그룹화한다.
    .groupby(['ORIGIN_AIRPORT', 'DESTINATION_AIRPORT'])
    # nb_flights : FLIGHT_NUMBER 의 집계수를 나타낸다.
    # 출발 - 도착 항공편을 기준으로 각 항공편마다의 사용률을 확인할 수 있다.
    .apply(lambda x: pd.Series({
        'nb_flights': len(x['FLIGHT_NUMBER'])
    }))
    # 행 인덱스 초기화.
    .reset_index()
    # rank_popular : ORIGIN_AIRPORT 를 기준으로 nb_flights 값을 비교하여 순위를 매긴다.
    .assign(
        rank_popular=lambda x:
            x.groupby(['ORIGIN_AIRPORT'])['nb_flights'].rank(method='first', ascending=False)
    )
    # 각 항공편마다 3위 안의 값들만 확인한다.
    [lambda x: x['rank_popular'] <= 3]
    # 행 인덱스를 초기화하고 nb_flights 열을 삭제한다. 
    .reset_index(drop=True)
    .drop(['nb_flights'], axis=1)
    # rank_popular 열의 내용을 top_숫자 의 형식으로 변경한다.
    .assign(rank_popular=lambda x: 'top_' + x['rank_popular'].astype(int).map(str))
    # rank_popular 을 기준으로 항공편을 줄 세운다.
    .pivot_table(
        columns='rank_popular',
        values='DESTINATION_AIRPORT',
        index='ORIGIN_AIRPORT',
        aggfunc=np.sum
    )
    # 행 인덱스를 초기화한다.
    # 결론적으로, 각 항공편마다 인기있는 노선을 확인한다.
    .rename_axis(None, axis=1)
    .reset_index()
)
result
```
