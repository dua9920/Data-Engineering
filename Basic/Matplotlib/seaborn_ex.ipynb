```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import seaborn as sns
```

```python
sns.set()
sns.set_style('whitegrid')
sns.set_color_codes()
```

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
sns.palplot(sns.color_palette('Blues'))
```

```python
sns.palplot(sns.dark_palette("muted purple", input="xkcd"))
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
print(list(zip(iris['sepal_length'], iris['sepal_width'])))
```

```python
sns.jointplot(x = 'sepal_length', y="sepal_width", data=iris)
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
plt.show()
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.show()
```

```python
#  
titanic_size = titanic.pivot_table(index="class", columns="sex", aggfunc="size")
titanic_size
```

```python
sns.heatmap(titanic_size, cmap=sns.light_palette('purple', as_cmap=True), annot=True, fmt='d')
plt.show()
```

```python
#   
flights_pass = flights.pivot("month", "year", "passengers")
print(flights_pass)
```

```python
sns.heatmap(flights_pass, annot=True, fmt='d', linewidths=1)
plt.show()
```
