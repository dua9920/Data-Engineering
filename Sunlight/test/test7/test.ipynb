```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import pyplot as plt, font_manager

font_path = "C:/Users/YEOM/AppData/Local/Microsoft/Windows/Fonts/AppleSDGothicNeoM.ttf"
font_name= font_manager.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)
```

```python
# 에너지 데이터 가져오기
energy = pd.read_csv("../../data/origin/energy.csv")
energy.head()
```

```python
# 에너지 데이터 개요 확인
energy.info()
```

```python
energy['time'] = pd.to_datetime(energy['time'], errors='coerce')
energy.head()
```

```python
# 2018년 ~ 2021년도 각 년도 데이터 뽑기
energy_2018 = energy[energy['time'].dt.year == 2018]
energy_2019 = energy[energy['time'].dt.year == 2019]
energy_2020 = energy[energy['time'].dt.year == 2020]
energy_2021 = energy[energy['time'].dt.year == 2021]
energy_2021
```

```python
# 각 년도 별 울산광역시 발전소 발전량
fig = plt.figure(figsize=(20, 7))
ax1 = fig.add_subplot(411)
plt.title('2018년 울산 발전소 발전량 추이', fontsize=17)
ax2 = fig.add_subplot(412)
plt.title('2019년 울산 발전소 발전량 추이', fontsize=17)
ax3 = fig.add_subplot(413)
plt.title('2020년 울산 발전소 발전량 추이', fontsize=17)
ax4 = fig.add_subplot(414)
plt.title('2021년 울산 발전소 발전량 추이', fontsize=17)
plt.subplots_adjust(left=0.125, bottom=0.1,  right=0.9, top=0.9, wspace=0.2, hspace=0.75) # 간격
ax1.plot(energy_2018['time'], energy_2018['ulsan'], color='lightseagreen')
ax2.plot(energy_2019['time'], energy_2019['ulsan'], color='lightcoral')
ax3.plot(energy_2020['time'], energy_2020['ulsan'], color='dodgerblue')
ax4.plot(energy_2021['time'], energy_2021['ulsan'], color='orange')
plt.show()
```
